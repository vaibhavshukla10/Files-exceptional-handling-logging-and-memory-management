1. What is the difference between interpreted and compiled languages ?

    Difference between Compiled and Interpreted Languages:

Compiled Languages:
- Code is translated into machine code before execution.
- Entire program is compiled at once.
- Faster execution.
- Examples: C, C++, Rust, Go.

Interpreted Languages:
- Code is executed line by line at runtime.
- No separate compilation step.
- Slower compared to compiled languages.
- Examples: Python, JavaScript, PHP, Ruby.

Key Difference:
- Compiled = first convert whole code, then run.
- Interpreted = convert and run at the same time.

2. What is exception handling in Python ?

   Exception handling in Python:

Exception handling is a way to handle runtime errors so that the program does not crash unexpectedly. It allows the program to continue execution even after an error occurs.

In Python, exception handling is done using:
- try
- except
- else
- finally

Example:

try:
    x = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
else:
    print("No error occurred")
finally:
    print("This block always executes")

Key Points:
- try block: code that may cause an error
- except block: handles the error
- else block: runs if no error occurs
- finally block: always runs regardless of error

3. What is the purpose of the finally block in exception handling?

    Purpose of the finally block in exception handling:

The finally block is used in exception handling to execute code that must run no matter what happens in the try and except blocks.

Key Purpose:
- It always executes whether an exception occurs or not.
- It is mainly used for cleanup operations.

Common Uses:
- Closing files
- Releasing resources (like database connections)
- Cleaning up memory or temporary data

Example:

try:
    x = 10 / 2
except ZeroDivisionError:
    print("Error occurred")
finally:
    print("This will always run")

Output:
- If error occurs → except runs + finally runs
- If no error → try runs + finally still runs

Conclusion:
The finally block ensures important cleanup code always executes.

15.  How does Python's garbage collection system work ?

     Python’s garbage collection system is responsible for automatically managing memory by removing objects that are no longer in use.

It works mainly using two mechanisms:

1. Reference Counting:
- Every object in Python has a reference count.
- When an object is created, its count increases.
- When references are removed, the count decreases.
- When the reference count becomes 0, memory is immediately freed.

2. Garbage Collector (Cycle Detection):
- Sometimes objects reference each other (circular references).
- Reference counting alone cannot remove them.
- Python uses a garbage collector (gc module) to detect and clean such cycles.

Working process:
- Python tracks all objects in memory.
- It identifies unreachable objects.
- It removes them to free memory.

3. Generational Approach:
- Objects are divided into generations (0, 1, 2).
- Newly created objects are checked more frequently.
- Old objects are checked less often.

6. Write a Python program that logs an error message to a log file when a division by zero exception occurs ?

```python
import logging

# Configure logging to write into a file
logging.basicConfig(
    filename="error_log.txt",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    num = int(input("Enter numerator: "))
    den = int(input("Enter denominator: "))

    result = num / den
    print("Result:", result)

except ZeroDivisionError as e:
    logging.error(f"Division by zero error occurred: {e}")
    print("Error: Cannot divide by zero (logged in file)")

except ValueError as e:
    logging.error(f"Invalid input error: {e}")
    print("Error: Invalid input (logged in file)")
```

18.  How would you implement a basic logging setup that logs to a file with rotation after 1MB ?

```python
import logging
from logging.handlers import RotatingFileHandler

# Create logger
logger = logging.getLogger("MyLogger")
logger.setLevel(logging.INFO)

# Create rotating file handler (1MB = 1,048,576 bytes)
handler = RotatingFileHandler(
    "app.log",
    maxBytes=1_000_000,   # 1MB
    backupCount=3         # keeps last 3 backup files
)

# Create format
formatter = logging.Formatter("%(asctime)s - %(levelname)s - %(message)s")
handler.setFormatter(formatter)

# Add handler to logger
logger.addHandler(handler)

# Generate logs
for i in range(100):
    logger.info(f"This is log message number {i}")
```
